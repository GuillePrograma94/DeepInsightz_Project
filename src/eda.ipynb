We will neutralise the negative (Return) products as long as there is a previous purchase in the same or previous year (max purchase data jan 2021) if there is no purchase value, instead of reducing the purchase count, we will eliminate the -1 value

```python
# Function to extract the information from the returns, by reducing the previous purchases, or eliminating the negative rows if there are no previous purchases made by the customer in the dataset we have

def process_returns(df):
    # Convert UNIDADES to numeric, replacing non-numeric values with NaN
    df['UNIDADES'] = pd.to_numeric(df['UNIDADES'], errors='coerce')
    
    # Remove rows where UNIDADES is NaN
    df = df.dropna(subset=['UNIDADES'])
    
    # Sort the dataframe by CLIENTE, ARTICULO, and FECHA
    df_sorted = df.sort_values(['CLIENTE', 'ARTICULO', 'FECHA'])
    
    # Separate positive (purchases) and negative (returns) transactions
    purchases = df_sorted[df_sorted['UNIDADES'] > 0].copy()
    returns = df_sorted[df_sorted['UNIDADES'] < 0].copy()
    
    # Process returns with progress bar
    total_returns = len(returns)
    for i, (_, return_row) in enumerate(tqdm(returns.iterrows(), total=total_returns, desc="Processing returns")):
        mask = (
            (purchases['CLIENTE'] == return_row['CLIENTE']) &
            (purchases['ARTICULO'] == return_row['ARTICULO']) &
            (purchases['FECHA'] <= return_row['FECHA']) &
            (purchases['UNIDADES'] > 0)
        )
        affected_purchases = purchases.loc[mask].sort_values('FECHA', ascending=False)
        
        units_to_return = abs(return_row['UNIDADES'])
        for idx, purchase in affected_purchases.iterrows():
            if units_to_return <= 0:
                break
            if purchase['UNIDADES'] >= units_to_return:
                purchases.loc[idx, 'UNIDADES'] -= units_to_return
                units_to_return = 0
            else:
                units_to_return -= purchase['UNIDADES']
                purchases.loc[idx, 'UNIDADES'] = 0
        
        # Print progress every 1000 iterations
        if (i + 1) % 1000 == 0:
            print(f"Processed {i + 1} / {total_returns} returns")
    
    # Remove rows where UNIDADES is 0 or less
    result = purchases[purchases['UNIDADES'] > 0].sort_values(['CLIENTE', 'ARTICULO', 'FECHA'])
    
    return result.reset_index(drop=True)
```
